```python
from datetime import datetime

def get_current_hour():
    now = datetime.now()
    current_hour = now.hour + 1
    return current_hour if current_hour <= 24 else 24

time = get_current_hour()
print(time)
```

```python
from datetime import datetime

# Function to retrieve the current hour in 24-hour format
def get_current_hour():
    now = datetime.now()
    return now.hour  # No need to add 1, just return now.hour for 0-23 format

hour_now = get_current_hour()
print(hour_now)
```
