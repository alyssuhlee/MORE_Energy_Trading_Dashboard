```python
from datetime import datetime
from openpyxl import load_workbook
import os
import openpyxl
import pandas as pd

# Current Rate = [(BCQSCPC x RateSCPC) + (BCQKSPC x RateKSPC) + (BCQEDC x RateEDC) - (TotalSSLoad-SummationBCQ) x DIPC HOURLY]/Total SS Load 
# Rates - Supplier Rates for the Month_July 2024

# DEFINING THE BASE DIRECTORY
base_directory_1 = r"C:\Users\aslee\OneDrive - MORE ELECTRIC AND POWER CORPORATION\MORE Energy Sourcing\013. Rate Analysis\005. 2024 Rate Analysis"

# -- GETTING THE CURRENT MONTH AND YEAR -- 
current_date = datetime.now()
current_month_numeric = current_date.strftime("%m")  # Month in numeric format (e.g., '01' for January)
current_month_name = current_date.strftime("%B")   # Full month name (e.g., 'January')
current_day = current_date.strftime("%d") # Day in numeric format (e.g., '26')
current_year = current_date.strftime("%Y")   # Year in 4-digit format (e.g., '2024')

#List of all month names
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
#Find index of current month
current_month_index = months.index(current_month_name)

# Initialize current_month_numeric_increment
current_month_numeric_increment = current_month_numeric # Default to current month if condition not met

if (current_day > '25') and (current_date.strftime("%m") == current_month_numeric):
    #Increment current month by converting to integer, adding 1, and formatting back to zero-padded string
    current_month_numeric_increment = '{:02d}'.format((int(current_month_numeric) % 12) + 1)
    #Calculate index of next month
    next_month_index = (current_month_index + 1) % 12
    current_month_name = months[next_month_index]

# -- CONSTRUCTING THE FOLDER NAME --
folder_prefix = current_month_numeric_increment.zfill(3)  # Zero-padded three-digit month number (e.g., '006' for June)
folder_name = f"{folder_prefix}. {current_month_name} {current_year}"

# -- CONSTRUCTING THE DIRECTORY PATH --
folder_path = os.path.join(base_directory_1, folder_name)

# -- CHECK IF THE FOLDER EXISTS AND OPEN IT -- 
if os.path.exists(folder_path):
    os.startfile(folder_path)
    print(f"Opened folder: {folder_path}")
else:
    print(f"Folder does not exist: {folder_path}")

# -- ADDITIONAL NEEDED VARIABLES FOR THE FOLDER NAME -- 
drs = "Daily Rate Simulation_"
extension = ".xlsx"

# -- FOLDER NAME (Example: Daily Rate Simulation_06252024) -- 
folder_name_2 = f"{drs}{current_month_numeric}{current_day}{current_year}{extension}"
folder_path_2 = os.path.join(folder_path, folder_name_2)

# -- FOR CREATING A NEW EXCEL FILE WITH THE NEEDED COLUMNS --
workbook = openpyxl.Workbook()
sheet = workbook.active
sheet['A1'] = 'HOUR' # done
sheet['B1'] = 'BCQ_SCPC' # done
sheet['C1'] = 'RATE_SCPC_B1' # done
sheet['D1'] = 'RATE_SCPC_B2' # done
sheet['E1'] = 'RATE_SCPC_AVE' # done
sheet['F1'] = 'BCQ_KSPC' # done
sheet['G1'] = 'RATE_KSPC_B1' # done
sheet['H1'] = 'RATE_KSPC_B2' # done
sheet['I1'] = 'RATE_KSPC_AVE' # done
sheet['J1'] = 'BCQ_EDC' # done
sheet['K1'] = 'RATE_EDC' # done
sheet['L1'] = 'TOTAL_SS_LOAD' # done
sheet['M1'] = 'TOTAL_BCQ' # done
sheet['N1'] = 'WESM_RATE' # done
sheet['O1'] = 'CURRENT_RATE'

workbook.save('current_rate.xlsx')
workbook.close()
print("Excel file 'current_rate.xlsx' created successfully.")

# File path for destination Excel file
destination_file_path = r'C:\Users\aslee\OneDrive - MORE ELECTRIC AND POWER CORPORATION\Desktop\DASHBOARD_FINAL\current_rate.xlsx'

# Load source workbook
source_wb = load_workbook(folder_path_2, data_only=True)  # Use data_only=True to get values instead of formulas
# Sheet name of the source workbook
source_sheet = source_wb['6. DAP Report']

# Load destination workbook
dest_wb = load_workbook(destination_file_path)

# Sheet name of the destination workbook
dest_sheet = dest_wb['Sheet']

# Define source and destination ranges
source_ranges = [('C11', 'C34'), ('G11', 'G34'), ('H11', 'H34'), ('I11', 'I34')]
dest_ranges = [('A2', 'A25'), ('B2', 'B25'), ('F2', 'F25'), ('J2', 'J25')]

# Function to copy values from source range to destination range
def copy_values(source_sheet, dest_sheet, source_range, dest_range):
    source_start, source_end = source_range
    dest_start, dest_end = dest_range
    
    # Copy values from source to destination
    source_values = []
    for row in source_sheet[source_start:source_end]:
        for cell in row:
            source_values.append(cell.value)
    
    dest_index = 0
    for row in dest_sheet[dest_start:dest_end]:
        for cell in row:
            if dest_index < len(source_values):
                cell.value = source_values[dest_index]
                dest_index += 1
            else:
                break

# Copy data from source to destination based on specified ranges
for i in range(len(source_ranges)):
    copy_values(source_sheet, dest_sheet, source_ranges[i], dest_ranges[i])

# Save the destination workbook
dest_wb.save(destination_file_path)

# Close workbooks
source_wb.close()
dest_wb.close()
print("Data transfer completed.")

# 2
value_B2 = dest_sheet['B2'].value
value_F2 = dest_sheet['F2'].value
value_J2 = dest_sheet['J2'].value

result2 = value_B2 + value_F2 + value_J2
dest_sheet['M2'] = result2

# 3
value_B3 = dest_sheet['B3'].value
value_F3 = dest_sheet['F3'].value
value_J3 = dest_sheet['J3'].value

result3 = value_B3 + value_F3 + value_J3
dest_sheet['M3'] = result3

# 4
value_B4 = dest_sheet['B4'].value
value_F4 = dest_sheet['F4'].value
value_J4 = dest_sheet['J4'].value

result4 = value_B4 + value_F4 + value_J4
dest_sheet['M4'] = result4

# 5
value_B5 = dest_sheet['B5'].value
value_F5 = dest_sheet['F5'].value
value_J5 = dest_sheet['J5'].value

result5 = value_B5 + value_F5 + value_J5
dest_sheet['M5'] = result5

# 6
value_B6 = dest_sheet['B6'].value
value_F6 = dest_sheet['F6'].value
value_J6 = dest_sheet['J6'].value

result6 = value_B6 + value_F6 + value_J6
dest_sheet['M6'] = result6

# 7
value_B7 = dest_sheet['B7'].value
value_F7 = dest_sheet['F7'].value
value_J7 = dest_sheet['J7'].value

result7 = value_B7 + value_F7 + value_J7
dest_sheet['M7'] = result7

# 8
value_B8 = dest_sheet['B8'].value
value_F8 = dest_sheet['F8'].value
value_J8 = dest_sheet['J8'].value

result8 = value_B8 + value_F8 + value_J8
dest_sheet['M8'] = result8

# 9
value_B9 = dest_sheet['B9'].value
value_F9 = dest_sheet['F9'].value
value_J9 = dest_sheet['J9'].value

result9 = value_B9 + value_F9 + value_J9
dest_sheet['M9'] = result9

# 10
value_B10 = dest_sheet['B10'].value
value_F10 = dest_sheet['F10'].value
value_J10 = dest_sheet['J10'].value

result10 = value_B10 + value_F10 + value_J10
dest_sheet['M10'] = result10

# 11
value_B11 = dest_sheet['B11'].value
value_F11 = dest_sheet['F11'].value
value_J11 = dest_sheet['J11'].value

result11 = value_B11 + value_F11 + value_J11
dest_sheet['M11'] = result11

# 12
value_B12 = dest_sheet['B12'].value
value_F12 = dest_sheet['F12'].value
value_J12 = dest_sheet['J12'].value

result12 = value_B12 + value_F12 + value_J12
dest_sheet['M12'] = result12

# 13
value_B13 = dest_sheet['B13'].value
value_F13 = dest_sheet['F13'].value
value_J13 = dest_sheet['J13'].value

result13 = value_B13 + value_F13 + value_J13
dest_sheet['M13'] = result13

# 14
value_B14 = dest_sheet['B14'].value
value_F14 = dest_sheet['F14'].value
value_J14 = dest_sheet['J14'].value

result14 = value_B14 + value_F14 + value_J14
dest_sheet['M14'] = result14

# 15
value_B15 = dest_sheet['B15'].value
value_F15 = dest_sheet['F15'].value
value_J15 = dest_sheet['J15'].value

result15 = value_B15 + value_F15 + value_J15
dest_sheet['M15'] = result15


# 16
value_B16 = dest_sheet['B16'].value
value_F16 = dest_sheet['F16'].value
value_J16 = dest_sheet['J16'].value

result16 = value_B16 + value_F16 + value_J16
dest_sheet['M16'] = result16

# 17
value_B17 = dest_sheet['B17'].value
value_F17 = dest_sheet['F17'].value
value_J17 = dest_sheet['J17'].value

result17 = value_B17 + value_F17 + value_J17 
dest_sheet['M17'] = result17

# 18
value_B18 = dest_sheet['B18'].value
value_F18 = dest_sheet['F18'].value
value_J18 = dest_sheet['J18'].value

result18 = value_B18 + value_F18 + value_J18
dest_sheet['M18'] = result18

# 19
value_B19 = dest_sheet['B19'].value
value_F19 = dest_sheet['F19'].value
value_J19 = dest_sheet['J19'].value

result19 = value_B19 + value_F19 + value_J19
dest_sheet['M19'] = result19

# 20
value_B20 = dest_sheet['B20'].value
value_F20 = dest_sheet['F20'].value
value_J20 = dest_sheet['J20'].value

result20 = value_B20 + value_F20 + value_J20
dest_sheet['M20'] = result20

# 21
value_B21 = dest_sheet['B21'].value
value_F21 = dest_sheet['F21'].value
value_J21 = dest_sheet['J21'].value

result21 = value_B21 + value_F21 + value_J21
dest_sheet['M21'] = result21

# 22
value_B22 = dest_sheet['B22'].value
value_F22 = dest_sheet['F22'].value
value_J22 = dest_sheet['J22'].value

result22 = value_B22 + value_F22 + value_J22
dest_sheet['M22'] = result22

# 23
value_B23 = dest_sheet['B23'].value
value_F23 = dest_sheet['F23'].value
value_J23 = dest_sheet['J23'].value

result23 = value_B23 + value_F23 + value_J23
dest_sheet['M23'] = result23

# 24
value_B24 = dest_sheet['B24'].value
value_F24 = dest_sheet['F24'].value
value_J24 = dest_sheet['J24'].value

result24 = value_B24 + value_F24 + value_J24
dest_sheet['M24'] = result24

# 25
value_B25 = dest_sheet['B25'].value
value_F25 = dest_sheet['F25'].value
value_J25 = dest_sheet['J25'].value

result25 = value_B25 + value_F25 + value_J25
dest_sheet['M25'] = result25

# Save the workbook back to file
dest_wb.save(destination_file_path)

# COLUMNS C, E, G (RATE) VAT EXCLUSIVE CHARGES
# Example: Supplier Rates for the Month_July 2024
srftm = "Supplier Rates for the Month_"
supplier_rates_folder = f"{srftm}{current_month_name} {current_year}"
supplier_rates_path = os.path.join(folder_path, supplier_rates_folder)

print(supplier_rates_path)
```
